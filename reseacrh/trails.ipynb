```python
import cv2
from ultralytics import YOLO, solutions
```

```python
model = YOLO('yolov8n.pt')
```

```python
path = r'c:\Users\91623\Downloads\run.mp4'
cap = cv2.VideoCapture(path)
```

```python
assert cap.isOpened()

w,h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
line_points = [(20, 400), (1080, 400)]

classes_to_count = [0,2]
vidoe_writer = cv2.VideoWriter("object_counting_output.avi", cv2.VideoWriter_fourcc(*"mp4v"), fps, (w,h))
```

```python
counter = solutions.ObjectCounter(
    view_img=True,  # Display the image during processing
    reg_pts=line_points,  # Region of interest points
    classes_names=model.names,  # Class names from the YOLO model
    draw_tracks=True,  # Draw tracking lines for objects
    line_thickness=2,  # Thickness of the lines drawn
)
```

```python
while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Video frame is empty or video processing has been succeessfully completed.")
        break

    tracks = model.track(im0, persist= True, show= False, classes=classes_to_count)

    im0 = counter.start_counting(im0, tracks)
    vidoe_writer.write(im0)

cap.release()
vidoe_writer.release()
cv2.destroyAllWindows()
```
